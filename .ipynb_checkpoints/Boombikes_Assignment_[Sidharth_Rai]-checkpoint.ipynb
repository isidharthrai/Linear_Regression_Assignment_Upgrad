# Boombikes - Demand analysis and prediction assignment
> Multiple Linear Regression Model for the prediction of demand for shared bikes.

### Import required libraries

```python
# Supress Warnings
import warnings
warnings.filterwarnings('ignore')

#import required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

import statsmodels.api as sm  

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

### Load Dataset

```python
#import dataset
boombikes = pd.read_csv('Dataset/day.csv')
```

```python
#print the head
boombikes.head()
```

```python
boombikes.info()
```

### Exploratory Data Analysis 
> Categorizing data into Continuous, Categorical data

```python
#data preparation, visualising the values in data to find the categorical variables

for x in boombikes.columns:
    print(f'Values in {x}\n',boombikes[x].value_counts())
```

> Classifying the columns on basis of datatypes </br>
>> Binary Types : Year, holiday, workingday </br>
>> Categorical Types: Season, Month, weekday, Weather </br>
>> Numeric types: temp, atemp, Humidity, windspeed, Count

```python
#selecting the required columns only
selected_cols = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit', 'temp', 'atemp', 
            'hum', 'windspeed','cnt']

df = boombikes[selected_cols]
df.head()
```

```python
df.columns
```

```python
#applying univariate & bivariate analysis on continuous variables, ignoring the categorical ones

cont_vars=['temp','atemp','hum','windspeed',"cnt"]

data = df[cont_vars] #segmented new dataset
print(data.describe())

sns.pairplot(df, vars=['temp','atemp','hum','windspeed',"cnt"])
plt.show()
```

```python
# from above graphs we can say that temp and atemp have a relationship
#let check the correlation
plt.figure(figsize = (16, 10))
sns.heatmap(df.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
#correlation between temp and atemp is 0.99 which is almost 1
#Let us drop temp and consider atemp

df.drop(['temp'],axis=1,inplace=True)
df.head()
```

```python
#from data we can see that: season,yr,mnth,holiday,weekday,workingday,weathersit all are categorical variables
#We will replace season,weekday and weathersit with appropriate values

df['season'].replace({1:"Spring",2:"Summer",3:"Fall",4:"Winter"},inplace=True)
df['weathersit'].replace({1:"Clear",2:"Mist",3:"Rain",4:'Severe'},inplace=True)
df['weekday'].replace({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday"},inplace=True)

df.head(10)
```

```python
#Visualising categorical Variables to understand data better
plt.figure(figsize=(30, 15))
plt.subplot(3,3,1)
sns.boxplot(x = 'season', y = 'cnt', data = df)#yr','mnth','workingday','weathersit','weekday'
plt.subplot(3,3,2)
sns.boxplot(x = 'yr', y = 'cnt', data = df)
plt.subplot(3,3,3)
sns.boxplot(x = 'mnth', y = 'cnt', data = df)
plt.subplot(3,3,4)
sns.boxplot(x = 'workingday', y = 'cnt', data = df)
plt.subplot(3,3,5)
sns.boxplot(x = 'weathersit', y = 'cnt', data = df)
plt.subplot(3,3,6)
sns.boxplot(x = 'weekday', y = 'cnt', data = df)
plt.subplot(3,3,7)
sns.boxplot(x = 'holiday', y = 'cnt', data = df)
plt.show()
```

### Creating Dummy Vars

```python
df.info()
```

```python
#Convert variables to object type
df['mnth']=df['mnth'].astype(object)
df['season']=df['season'].astype(object)
df['weathersit']=df['weathersit'].astype(object)
df['weekday']=df['weekday'].astype(object)
df.info()
```

```python
#creating dummies for categorical cols
#drop the first column as it will be significant to understand the category refering to.
season = pd.get_dummies(df['season'], dtype='int', drop_first = True)
month = pd.get_dummies(df['mnth'], dtype='int', drop_first = True)
weekday = pd.get_dummies(df['weekday'], dtype='int', drop_first = True)
weathersit = pd.get_dummies(df['weathersit'], dtype='int', drop_first = True)

print(season.head())
print(month.head())
print(weekday.head())
print(weathersit.head())
```

```python
df=pd.concat([df,season],axis=1)
df=pd.concat([df,month],axis=1)
df=pd.concat([df,weekday],axis=1)
df=pd.concat([df,weathersit],axis=1)
df.info()
```

```python
#drop the original columns from df as it becomes irrelevant
df.drop(['season', 'mnth','weekday','weathersit'], axis = 1, inplace = True)

#get new info
df.info()
```

### Splitting the dataset

```python
# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
df_train.columns
```

```python
#scale the numeric columns between 0 and 1
scaler=MinMaxScaler()

num_vars=['atemp','hum','windspeed','cnt']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.describe()
```

```python
#creating predictive variable
y_train = df_train.pop('cnt')
X_train = df_train
```

### Feature Selection & Linear Regression

```python
# Convert feature names to strings
X_train.columns = X_train.columns.astype(str)
```

```python
# Initialize and fit the linear regression model
lm = LinearRegression()
lm.fit(X_train, y_train)

# Fit the RFE model on the training data
rfe = RFE(estimator=lm, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

### Building model using statsmodel, for the detailed statistics

```python
X_train_rfe1 = sm.add_constant(X_train_rfe)

lm = sm.OLS(y_train,X_train_rfe1).fit()

print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#column workingday has very high VIF so dropping it.
X_train_rfe=X_train_rfe.drop(['workingday'],axis=1)

X_train_rfe1 = sm.add_constant(X_train_rfe)

lm1 = sm.OLS(y_train,X_train_rfe1).fit()

print(lm1.summary())
```
